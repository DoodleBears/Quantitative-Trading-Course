```python
import seaborn as sns
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
from datetime import datetime
import openpyxl
import talib
import numpy as np
```

```python
# 渲染图设定
plt.rcParams['figure.figsize'] = (12.0, 8.0)
```

```python
start=datetime(2020,1,1)
end =datetime.today()
tw_2317=pdr.get_data_yahoo('2317.tw',start,end) # 鸿海
tw_2330=pdr.get_data_yahoo('2330.tw',start,end) # 台積電
tw_2881=pdr.get_data_yahoo('2881.tw',start,end) # 富邦金控
tw_2882=pdr.get_data_yahoo('2882.tw',start,end) # 国泰金控
# print(Aapl_usa.head(5))
# print(Aapl_usa.tail(5)) 
# print(Aapl_usa.shape)
```

```python
stock=pd.DataFrame({
        '2317.tw':tw_2317['Adj Close'],
        '2330.tw':tw_2330['Adj Close'],
        '2881.tw':tw_2881['Adj Close'],
        '2882.tw':tw_2882['Adj Close']
    })
# tw_2317['Adj Close'].plot(label='2317.tw') 
# tw_2881['Adj Close'].plot(label='2881.tw') 
# tw_2882['Adj Close'].plot(label='2882.tw') 
stock.plot(grid=True)
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.legend()
plt.show()
```

```python
#計算移動平均值
MA_timeperiod = 5
stock_SMA = pd.DataFrame({
    'SMA%s-2317.tw' %(MA_timeperiod):talib.SMA(tw_2317['Adj Close'], MA_timeperiod),
    # 'SMA%s-2330.tw' %(MA_timeperiod):talib.SMA(tw_2330['Adj Close'], MA_timeperiod),
    'SMA%s-2881.tw' %(MA_timeperiod):talib.SMA(tw_2881['Adj Close'], MA_timeperiod),
    'SMA%s-2882.tw' %(MA_timeperiod):talib.SMA(tw_2882['Adj Close'], MA_timeperiod),
})
stock_SMA.plot(grid=True)
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.legend()
plt.show()
```

```python
#計算移動平均值
MA_timeperiod = 20
stock_KAMA = pd.DataFrame({
    'KAMA%s-2317.tw' %(MA_timeperiod):talib.KAMA(tw_2317['Adj Close'], MA_timeperiod),
    # 'KAMA%s-2330.tw' %(MA_timeperiod):talib.KAMA(tw_2330['Adj Close'], MA_timeperiod),
    'KAMA%s-2881.tw' %(MA_timeperiod):talib.KAMA(tw_2881['Adj Close'], MA_timeperiod),
    'KAMA%s-2882.tw' %(MA_timeperiod):talib.KAMA(tw_2882['Adj Close'], MA_timeperiod),
})
stock_KAMA.plot(grid=True)
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.legend()
plt.show()
```
